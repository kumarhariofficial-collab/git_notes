## While Loop

```python
password = ""
while password != "login@123":
    password = input("Enter the password:")
    if password != "login@123":
        print("Incorrect password. try again")
    print("Access granted")
```

```python
number = -1

while number <= 0:
    number = input("Enter the positive number")
    if number <= 0:
        print("Invalid! please enter a positive number:")
    print("Thank you! you have entered:" ,{number})
```

```python
number = -1

while number <= 0:
    number = int(input("Enter the positive number"))
    if number <= 0:
        print("Invalid! please enter a positive number:")
    print("Thank you! you have entered:" ,number)
```

## UDF - No Loop/If

```python
def format_name(name):
    return name.strip().title()

print(format_name("  harikumar p  "))
```

```python
def get_percentage(value, total):
    return (value / total) * 100

print(get_percentage(45, 50))
```

## UDF with For Loop

```python
def create_skill(skill_list):
    result = ""
    for s in skill_list:
        result += s + " | "
    return result

print(create_skill(["SQL", "Salesforce", "Excel", "Python"]))
```

```python
def clean_list(items):
    for item in items:
        print(item.strip())

clean_list(["  Amazon ", "  Google  "])
```

## UDF with For Loop and If Condition

```python
def find_tool(tools, target):
    for t in tools:
        if t == target:
            return f"{target} is in your toolkit!"
    return "Tool not found."

print(find_tool(["Power BI", "Python", "SQL"], "Python"))
```

```python
def check_metrics(scores):
    for s in scores:
        if s > 90:
            print(f"Outstanding Score: {s}")
check_metrics([85, 92, 78, 95])
```
